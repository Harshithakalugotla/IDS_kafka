```python
import numpy as np
import matplotlib.pyplot as plt

# Algorithms
algorithms = [
    "IF + XGB",
    "KMeans + XGB",
    "PCA + XGB",
    "LOF + XGB",
    "RF + XGB"
]

# Metrics
accuracy = [0.9959, 0.9943, 0.9970, 0.9949, 0.9964]
precision = [0.9960, 0.9958, 0.9969, 0.9950, 0.9964]
recall = [0.9955, 0.9928, 0.9971, 0.9949, 0.9964]
f1 = [0.9959, 0.9943, 0.9970, 0.9949, 0.9964]

# X positions
x = np.arange(len(algorithms))
width = 0.2

plt.figure(figsize=(10,6))

# 🎨 Pastel colors
colors = ['#A8DADC', '#FFD6A5', '#BDB2FF', '#FFC6FF']

bars1 = plt.bar(x - 1.5*width, accuracy, width, label='Accuracy', color=colors[0])
bars2 = plt.bar(x - 0.5*width, precision, width, label='Precision', color=colors[1])
bars3 = plt.bar(x + 0.5*width, recall, width, label='Recall', color=colors[2])
bars4 = plt.bar(x + 1.5*width, f1, width, label='F1 Score', color=colors[3])

# Labels and title
plt.xlabel("Models")
plt.ylabel("Score")
plt.title("Performance Comparison of Hybrid Models")
plt.xticks(x, algorithms, rotation=30)

# 🔥 FIX: more headroom (no cut-off)
plt.ylim(0.99, 1.0025)

plt.legend()

# 🔥 FIX: staggered + diagonal labels (no overlap)
offsets = [0.0002, 0.0004, 0.0006, 0.0008]

for bars, offset in zip([bars1, bars2, bars3, bars4], offsets):
    for bar in bars:
        yval = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width()/2,
            yval + offset,
            f"{yval:.4f}",
            ha='center',
            va='bottom',
            rotation=45,
            fontsize=8
        )

# 🔥 FIX: prevent top clipping
plt.subplots_adjust(top=0.88)

plt.tight_layout()
plt.show()
```
